```python
import folium
import pandas as pd
```

```python
sales_data = pd.read_excel('绘图数据.xlsx',sheet_name='Sales')
distributors_data = pd.read_excel('绘图数据.xlsx',sheet_name='Distributors')
```

```python
sales_data.head(3)
```

```python
sales_data_list = list(zip(sales_data['Latitude'],sales_data['Longitude']))
distributors_data_list = list(zip(distributors_data['Latitude'],distributors_data['Longitude']))
```

```python
m = folium.Map(location=[45.372, -121.6972], zoom_start=12, tiles="OpenStreetMap")

# 绘制销售点的数据
for i in range(len(sales_data_list)):
    folium.Marker(
        location=sales_data_list[i],
        popup="城市{},所在州{},邮编{},周销量{}".format(sales_data['Destination City'][i],sales_data['Destination State'][i],sales_data['Destination Zip Code'][i],sales_data['Weight'][i]),
        icon=folium.Icon(icon="cloud"),
    ).add_to(m)

# 绘制工厂的位置
folium.Marker(
    location=(40.5365859,-79.9072774),
    popup="工厂地址{},邮编{}".format('Blawnox','15238'),
    icon=folium.Icon(color="red",icon="industry"),
).add_to(m)

# 绘制中转仓库的位置
for i in range(len(distributors_data_list)):
    folium.Marker(
        location=distributors_data_list[i],
        popup="地址{},邮编{}".format(distributors_data['Location'][i],distributors_data['Zip Code'][i]),
        icon=folium.Icon(color="orange",icon="warehouse-full"),
    ).add_to(m)

m
```

```python
m.save("locations.html")
```
